1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.
2. Plote o scatterplot de altura x peso da nova representação.

**Código do dataset:** women

**Dicas:**

- 1 polegada = 2,54 centímetros
- 1 libra = 0,45 quilos
- use o método .plot.scatter()

1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.

---



```python
from pydataset import data
import pandas as pd
import matplotlib.pyplot as plt
#matplotlib inline

# importando o dataset usando a função data("Código") do pydataset e atribuindo uma variavel para receber o dataset
df2 = data('women')

df2.head()
```

```python
# Construindo uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.
#df2.columns.values

df2['Altura'] = df2['height'] * 2.54
df2['Peso'] = df2['weight'] * 0.45

# Excluindo as colunas em inglês
df2.drop(columns = ['height', 'weight'], axis = 1, inplace = True)

# Vendo como ficou
df2.head()
```

2. Plote o scatterplot de altura x peso da nova representação.

---



```python
df2.plot.scatter('Altura', 'Peso', color = 'purple', s=15)
plt.title('Altura x peso de mulheres estadunidenses')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (Kg)')
```

**Definição do problema:** Você está trabalhando em um jornal e o editor

  1. Qual a empresa mais valiosa da lista da Forbes 2000?
  2. Construa uma tabela com as TOP 10 empresas mais lucrativas da

  3. Qual a média de valores de empresa das cinco categorias mais

**Código do dataset:** Forbes2000

**Dicas:**

1. Qual a empresa mais valiosa da lista da Forbes 2000?

---



```python
from pydataset import data
import pandas as pd
# Importando o dataset usando a função data("Código") do pydataset e atribuindo uma variavel para receber o dataset
datasetforbes = data('Forbes2000')
# Visualizando o dataset
datasetforbes
```

```python
# Agrupando as empresas pelos maiores marketvalues
maxmarketvalue = datasetforbes.groupby('marketvalue').max()
# Ordenando os marketvalue do maior para o menor
marketvalueordenado = maxmarketvalue.sort_values(by=['marketvalue'],ascending=False)
# Criando dataframe com as colunas name e marketvalue
marketvalue = ['name','marketvalue']
# Criando dataframe e filtrando as colunas name e marketvalue
maxmarketvalueordenado = marketvalueordenado.filter(items=marketvalue)
#Exibindo o dataframe com os nomes das Empresas e seus marketvalues em ordem decrescente
maxmarketvalueordenado
# A Empresa mais valiosa é a General Eletrics
```

2. Construa uma tabela com as TOP 10 empresas mais lucrativas da lista

---



```python
# Selecionando as 10 empresas mais lucrativas
top10 = datasetforbes.nlargest(10,"profits")
# Criando dataframe com as colunas name e profits
profits = ['name','profits']
# Criando dataframe e filtrando as colunas name e profits
top10filtrado = top10.filter(items=profits)
#Exibindo o dataframe com os nomes das Empresas e seus profits
top10filtrado
```

3. Qual a média de valores de empresa das cinco categorias mais valiosas?

---



```python
# Agrupando empresas por categoria e somando para verificar quais as mais valiosas
valiosas = datasetforbes.groupby('category').sum()
# Ordenando os marketvalue somados do maior para o menor
valiosasordenado = valiosas.sort_values(by=['marketvalue'],ascending=False)
# Criando dataframe com as colunas name e marketvalue
marketvalue2 = ['name','marketvalue']
# Criando dataframe e filtrando as colunas name e marketvalue
valiosasfiltrado = valiosasordenado.filter(items=marketvalue2)
#Selecionando o nome das 5 categorias mais valiosas
nomeempresas = valiosasfiltrado.index.values
top5categorias = nomeempresas[0:5]
print(top5categorias)

```

```python

datacategorias = datasetforbes[datasetforbes['category'].isin(top5categorias)]
# Selecionar apenas as colunas numéricas
numeric_columns = ['category','sales', 'profits', 'assets', 'marketvalue']
# Criando dataframe e filtrando as colunas
mediafiltrado = datacategorias.filter(items=numeric_columns)
# Agrupando empresas por categoria e calculando a média
mediacategoria = mediafiltrado.groupby('category').mean()
# Ordenando as médias do marketvalue do maior para o menor
mediaordenado = mediacategoria.sort_values(by=['marketvalue'],ascending=False)
mediaordenado



```
